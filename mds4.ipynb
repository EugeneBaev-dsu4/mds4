## Домашнее задание "Математический анализ. Производная"

Уровень 0:
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```python
import numpy as np
from scipy.misc import derivative
```

```python
def func(x):
    return (x**5) + 4*np.sin(2*x) + np.cos(3*x+3)
print('первая производная: ', derivative(func, 1.0, dx=1e-6, n=1))
print('вторая производная: ', derivative(func, 1.0, dx=1e-6, n=2))
```

Уровень 1:
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
def func_2(x):
    return np.sin(2*x+1)**5
def func_der1(x):
    return derivative(func_2, x, dx=1e-6, n=1)
def func_der2(x):
    return derivative(func_2, x, dx=1e-6, n=2)
```

```python
x_range = np.linspace(-5,5,1000) 
fd1_range = list(map(func_der1, x_range))
fd2_range = list(map(func_der2, x_range))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(x_range,fd1_range)
plt.plot(x_range,fd2_range)
plt.show()
```
